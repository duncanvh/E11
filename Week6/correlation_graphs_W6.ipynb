```python
import math
import numpy as np 
import matplotlib.pyplot as plt
import csv
import pandas as pd
from scipy.stats import pearsonr
```

```python
outdata = pd.read_csv('busstopdata.csv')
indata = pd.read_csv('insidedata.csv')
```

```python
display(indata)
```

```python
display(outdata)
```

```python
temp_out  = outdata.loc[1:,'BM680 Temp']
hum_out = outdata.loc[1:,'BM680 Humidity']
plt.scatter(temp_out,hum_out, alpha = 0.5)
plt.ylabel('Humidity')
plt.xlabel('Temperature')
plt.title('Humidity Vs Temperature (outside)')
plt.grid(True)
plt.show()
r_pearson , p_value = pearsonr(temp_out,hum_out)
if r_pearson >= 0.5:
    print(f' the r value: {r_pearson} means a positive strong correlation ')
elif 0<r_pearson<0.5:
    print(f'the r value: {r_pearson} means a weak positive correlation')
elif -0.5<r_pearson<0:
    print(f'the r value:{r_pearson} means a weak negative correlation')
else: r_pearson<=-0.5
print(f'the r value: {r_pearson} means a strong negative correlation')
if p_value<0.05:
    print(f'the p value:{p_value} means statistically significant correlation')
    
#there are many hidden variables that could affect these graphs such as, the time of day, the air flow
# and even weather condiditons could affect the outcomes of both graphs.
```
